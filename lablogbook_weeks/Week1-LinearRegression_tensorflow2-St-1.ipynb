# Linear Regression Example using tensorflow
Linear regression implementation with TensorFlow v2 library.

This example is using a low-level approach to better understand all mechanics behind the training process.


```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
rng = np.random
```

```python
# Parameters
learning_rate = 0.01
training_steps = 1000
display_step = 50
```

```python
# Just run the next cell using 'Ctrl' + 'Enter'
```

```python
# Training Data.

X = np.array([3.3,4.4,5.5,6.71,6.93,4.168,9.779,6.182,7.59,2.167,
              7.042,10.791,5.313,7.997,5.654,9.27,3.1])

Y = np.array([1.7,2.76,2.09,3.19,1.694,1.573,3.366,2.596,2.53,1.221,
              2.827,3.465,1.65,2.904,2.42,2.94,1.3])

n_samples = X.shape[0]
```

```python
print(X, '\n', X.shape)
```

```python
print(Y, '\n', Y.shape)
```

```python
# Weight and Bias, initialised randomly

W = tf.Variable(rng.randn(), name="weight")
b = tf.Variable(rng.randn(), name="bias")

print(W.numpy(), '\n', b.numpy())
```

```python
# Linear regression (Wx + b)

def linear_regression(x):
    return W * x + b
```

```python
# Mean square error
def mean_square(y_pred, y_true):
    return tf.reduce_sum(tf.pow(y_pred-y_true, 2)) / (2 * n_samples)
```

```python
# Stochastic Gradient Descent Optimizer
optimzer = tf.optimizers.SGD(learning_rate)
```

```python
# Adding a question mark () after the function name lets you view the latest documentation. 
# This is a handy tool.
```

```python
tf.optimizers.SGD?
```

```python
# Optimisation process

def run_optimization():
    
    # Wrap computation inside a GradientTape for automatic differentiation
    with tf.GradientTape() as g:
        pred = linear_regression(X)
        loss = mean_square(pred, Y)
        
    # Compute gradients
    gradients = g.gradient(loss, [W, b])
    
    # Update W aand b following gradients 
    optimizer.apply_gradients(zip(gradients, [W, b]))
```

```python
# Define the optimizer
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
# Run training for the given number of steps

for step in range(1, training_steps + 1):
    
    # Run the optimisation to update W and b values
    run_optimization()
    
    if step % display_step == 0:
        pred = linear_regression(X)
        loss = mean_square(pred, Y)
        print("step: %i, loss: %f, W: %f, b: %f" % (step, loss, W.numpy(), b.numpy()))
```

```python
# Plot the Graph

plt.plot(X, Y, 'ro', label='Original data')
plt.plot(X, np.array(W * X + b), label='Fitted line')
plt.legend()
plt.show()
```

```python
# Plot the original data
plt.figure(figsize=(8, 6))  # Adjust the figure size
plt.plot(X, Y, 'ro', label='Original data', markersize=8)  # Adjust marker size

# Plot the fitted line
plt.plot(X, np.array(W * X + b), 'b-', label='Fitted line', linewidth=2)  # Adjust line color and width

# Customize the plot
plt.title('Linear Regression', fontsize=16)  # Set the title
plt.xlabel('X', fontsize=14)  # Set the x-axis label
plt.ylabel('Y', fontsize=14)  # Set the y-axis label
plt.legend(fontsize=12)  # Show the legend
plt.grid(True)  # Show the grid

# Display the plot
plt.show()
```
